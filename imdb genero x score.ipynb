### tratamento

```python
# retira as colunas com dados faltantes
novos_dados = dados.dropna()
```

```python
## quantidade de linhas "perdidas"
print(len(dados) - len(novos_dados))
```

```python
novos_dados.index #possivel observar que existe alguns indices faltantes
```

```python
novos_dados.index = range(0,3756) #cria novos indices
novos_dados
```

```python
#função para realizar a contagem de generos por filme
def contador_generos(lista_generos):
    qtd_generos = [] #cria uma lista
    for i in range(0,len(lista_generos)):
        qtd = len(lista_generos[i].split('|'))
        qtd_generos.append(qtd)
        
    return qtd_generos
    
```

```python
#cria uma nova coluna contendo a quantidade de generos
lista_generos = novos_dados.genres
novos_dados['qtd_generos'] = contador_generos(lista_generos)
```

```python
## fim do tratamento nos dados
novos_dados.head(10)
```

### análise descritiva

```python
%matplotlib inline
#mostra a distribuição dos scores em relação a quantidade de gêneros em cada filme
sns.pairplot(novos_dados, x_vars=["qtd_generos"], y_vars ='imdb_score', size=7, kind='reg')
```
